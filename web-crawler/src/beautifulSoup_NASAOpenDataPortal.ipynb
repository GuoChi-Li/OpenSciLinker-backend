# Import modules

```python
import requests
from bs4 import BeautifulSoup
import json
import datetime
```

```python
year = datetime.date.today().year
month = datetime.date.today().month
day = datetime.date.today().day
```

# Set the url

```python
def setURL(url, num_page=-1):
    if num_page == -1:
        response = requests.get(str(url))
        soup = BeautifulSoup(response.text, 'html.parser')
        return soup
    else:
        response = requests.get(str(url) + str(num_page))
        soup = BeautifulSoup(response.text, 'html.parser')
        return soup
```

# Find the project title, description and URL

```python
num_page = 4
# num_page = 3 # for test
```

```python
for page_index in range(1, num_page):
    soup = setURL('https://data.nasa.gov/browse?limitTo=datasets&page=', page_index)

    title_body = soup.find_all('h2', class_='browse2-result-name')
    link_body = soup.find_all('a', class_='browse2-result-name-link')

    each_result = soup.find_all('div', class_="browse2-result-content clearfix")
    date_body = soup.find_all('span', class_='dateLocalize')
    for index in range(len(title_body)):
        print(index) # index
        print(title_body[index].text) # title

        if (str(each_result[index].find('div', class_='browse2-result-description collapsible-content'))) == "None":
            # description = "None"
            print("None")
        else:
            print(each_result[index].find('div', class_='browse2-result-description collapsible-content').text)

        print(link_body[index].get('href')) # link
        if (str(each_result[index].find('a', itemprop="keywords"))) == "None":
            print("None")
        else:
            print(each_result[index].find('a', itemprop="keywords").text)
        print(date_body[index].text) # date
```

# Saving the JSON file

```python
json_data = []

for page_index in range(1, num_page):
    soup = setURL('https://data.nasa.gov/browse?limitTo=datasets&page=', page_index)

    title_body = soup.find_all('h2', class_='browse2-result-name')
    link_body = soup.find_all('a', class_='browse2-result-name-link')

    each_result = soup.find_all('div', class_="browse2-result-content clearfix")
    date_body = soup.find_all('span', class_='dateLocalize')
    for index in range(len(title_body)):
        
        if (str(each_result[index].find('div', class_='browse2-result-description collapsible-content'))) == "None":
            # description = "None"
            des = "None"
        else:
            des = str(each_result[index].find('div', class_='browse2-result-description collapsible-content').text)

        if (str(each_result[index].find('a', itemprop="keywords"))) == "None":
            key = "None"
        else:
            key = str(each_result[index].find('a', itemprop="keywords").text)

        json_data.append({"title":title_body[index].text.strip(),'Project URL':link_body[index].get('href'), 'description': des.strip(), 'keywords':key.strip(), 'date': date_body[index].text.strip()})
```

```python
print(json_data[:5])
```

```python
json_data = json.dumps(json_data, indent=4)  # 'indent' for pretty formatting

# Specify the file path where you want to save the JSON data
file_path = "./../data/NASAOpenDataPortal/" + str(year) + str(month) + str(day) + "_data.json"

# Write the JSON string to the file
with open(file_path, 'w', encoding='utf-8') as json_file:
    json_file.write(json_data)
```
