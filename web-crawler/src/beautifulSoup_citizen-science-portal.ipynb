# Import modules

```python
import requests
from bs4 import BeautifulSoup
import json
import datetime
```

```python
year = datetime.date.today().year
month = datetime.date.today().month
day = datetime.date.today().day
```

# Set the url

```python
response = requests.get('https://science.gc.ca/site/science/en/citizen-science-portal')
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# print(soup.prettify())
```

# Find the project title, description and location

```python
tbody = soup.find_all('td')
tbody = tbody[3:]
for index, element in enumerate(tbody):
    # print(element)
    if index%3==0:
        # print(index)
        print(element.find('strong').text)
        try:
            print(element.find('a').get('href'))
        except:
            pass
    if index%3==1:
        print('description: ', element.text)
    if index%3==2:
        print('location: \n', element.text)
    print('-'*30)
```

# Saving the JSON file

```python
json_data = []
tbody = soup.find_all('td')
tbody = tbody[3:] # remove the first three elements
for index in range(len(tbody)):
    # print(element)
    if index%3==0:
        # print(index)
        try:
            # If the project has a URL
            json_data.append({"title":tbody[index].find('strong').text.strip(),'Project URL':tbody[index].find('a').get('href'), 'description': tbody[index+1].text.strip(), 'location': tbody[index+2].text.strip()})
        except:
            # If the project does not have a URL
            json_data.append({"title":tbody[index].find('strong').text.strip(),'Project URL':'None', 'description': tbody[index+1].text.strip(), 'location': tbody[index+2].text.strip()})
print(json_data[:5])
```

```python
json_data = json.dumps(json_data, indent=4)  # 'indent' for pretty formatting

# Specify the file path where you want to save the JSON data
file_path = "./../data/GovernmentofCanadaCitizenSciencePortal/" + str(year) + str(month) + str(day) + "_data.json"

# Write the JSON string to the file
with open(file_path, 'w', encoding='utf-8') as json_file:
    json_file.write(json_data)
```
