```python
import json
import datetime
```

If a project doesn't have these key, use empty string "".

- title -> project_name
- description -> project_description
- project URL -> project_url_external
- keywords -> sub_tag
- category {GitHub_repo, Data, Project}
- created_date (YYYY-MM-DD)
- owner -> publisher


```python
file_path = './../data/NASAOpenDataPortal/2023107_data.json'
save_path = './../data/NASAOpenDataPortal/clean/2023107_data.json'
```

```python
json_data = []
```

```python
# Open the JSON file for reading
with open(file_path, 'r') as file:
    # Load the JSON data
    data = json.load(file)
    json_len = len(data)

    for index in range(json_len):
        project_name = data[index]['title']
        project_description = data[index]['description']
        project_url_external = data[index]['Project URL']
        sub_tag =  data[index]['keywords']
        category = 'Project'
        publisher = ""

        # formatting the date, January 19 2021
        dateString = data[index]['date']
        dateFormatter = "%B %d %Y"
        datetime_object = datetime.datetime.strptime(dateString, dateFormatter)
        created_date = str(datetime_object.year) + '-' + str(datetime_object.month) + '-' + str(datetime_object.day)
        
        # print(index)
        print(project_name)
        print(project_description)
        print(project_url_external)
        print(sub_tag)
        print(category)
        print(created_date)
        print(publisher)

        json_data.append({"project_name": project_name, "project_description": project_description, "project_url_external": project_url_external, "sub_tag": sub_tag, "category": category, "created_date": created_date, "publisher": publisher})

```

```python
json_data = json.dumps(json_data, indent=4)  # 'indent' for pretty formatting

# Write the JSON string to the file
with open(save_path, 'w', encoding='utf-8') as json_file:
    json_file.write(json_data)
```
