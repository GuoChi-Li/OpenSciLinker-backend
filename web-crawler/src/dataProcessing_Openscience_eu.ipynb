```python
import json
import datetime
```

If a project doesn't have these key, use empty string "".

- title -> project_name
- description -> project_description
- project URL -> project_url_external
- keywords -> sub_tag
- category {GitHub_repo, Data, Project}
- created_date (YYYY-MM-DD)
- owner -> publisher


```python
file_path = './../data/Openscience_eu/2023107_data.json'
save_path = './../data/Openscience_eu/clean/2023107_data.json'
```

```python
json_data = []
```

```python
# Open the JSON file for reading
with open(file_path, 'r') as file:
    # Load the JSON data
    data = json.load(file)
    json_len = len(data)

    for index in range(json_len):
        # print('-'*50)
        project_name = data[index]['title']
        project_description = data[index]['description']
        project_url_external = data[index]['Project URL']
        sub_tag = ''
        category = 'Project'
        created_date = ''
        publisher = ""
        location = data[index]['location']

        
        # print('before print')
        # print(index)
        print(project_name)
        print(project_description)
        print(project_url_external)
        print(sub_tag)
        print(category)
        print(created_date)
        print(publisher)

        json_data.append({"project_name": project_name, "project_description": project_description, "project_url_external": project_url_external, "sub_tag": sub_tag, "category": category, "created_date": created_date, "publisher": publisher, "location": location})

```

```python
json_data = json.dumps(json_data, indent=4)  # 'indent' for pretty formatting

# Write the JSON string to the file
with open(save_path, 'w', encoding='utf-8') as json_file:
    json_file.write(json_data)
```
